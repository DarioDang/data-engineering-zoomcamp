#### Handling Data Schema 

```python
import pandas as pd

df = pd.read_parquet("data/green_tripdata_2025-11.parquet")

# take first 100 rows
df = df.head(100)

# convert datetime
df["lpep_pickup_datetime"] = pd.to_datetime(df["lpep_pickup_datetime"])
df["lpep_dropoff_datetime"] = pd.to_datetime(df["lpep_dropoff_datetime"])


parse_dates = [
    "lpep_pickup_datetime",
    "lpep_dropoff_datetime"
]

# enforce dtypes after reading
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64",
}
df = df.astype(dtype)

df.head()
```

#### Ingest Data To Postgres 

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace') # n = 0 jsut create the schema
```

```python
from sqlalchemy import create_engine
import pyarrow.parquet as pq
from tqdm.auto import tqdm

parquet_file = pq.ParquetFile("data/green_tripdata_2025-11.parquet")

for batch in tqdm(
    parquet_file.iter_batches(batch_size=100_000),
    total=parquet_file.metadata.num_rows // 100_000 + 1,
    desc="Ingesting parquet"
):
    df = batch.to_pandas()
    df = df.astype(dtype)

    df.to_sql(
        name="green_tripdata_2025_11",
        con=engine,
        if_exists="append",
        index=False,
        method="multi"
    )
```

### Add the Zone Dataset to Postgres

```python
zones = pd.read_csv("data/taxi_zone_lookup.csv")

# normalize column names to match DB style
zones.columns = [c.lower() for c in zones.columns]

# define schema via pandas dtypes (similar idea to taxi trips)
zones_dtype = {
    "locationid": "Int64",
    "borough": "string",
    "zone": "string",
    "service_zone": "string"
}

zones = zones.astype(zones_dtype)

zones.head()
```

```python
# Generate schema for zones df
print(
    pd.io.sql.get_schema(
        zones,
        name="taxi_zone_lookup",
        con=engine
    )
)
```

```python
zones.to_sql(
    name="taxi_zone_lookup",
    con=engine,
    schema="public",
    if_exists="replace",
    index=False
)
```
